#### **[task 6, 1б.]**
Познакомимся с рынком труда России. Для этого допишите функцию info_count.

```python
def info_count(df, col):
    df = df.groupby(by=[col]).count()
    df = df.sort_values(by=['region'])
    
    df['region'] = df['region'] / df['region'].sum()
    g = df['region']
    
    g.plot(kind='barh')
    for index, value in enumerate(g):
        plt.text(value, index, str(round(value*100,1))+'%')
    plt.grid(True)
    plt.show()
```

```python
plot_cols = ['town_type',  
            'professional_group',
            'finished_education', 
            'sex', 
            'employed', 'industry']

for col in plot_cols:
    info_count(df, col)
```

```python
def info_wage_median(df, col):
    
    df = df.groupby(by=[col]).median()
    df = df.sort_values(by=['wage'])
    df['wage'] = df['wage']/1000
    g = df['wage']

    fig, ax = plt.subplots(figsize = (12,6))
    g.plot(kind='barh')
    for index, value in enumerate(g):
        plt.text(value, index, str(int(value)))
    plt.grid(True)
    plt.xlabel('тыс. руб.')
    plt.show()

for col in plot_cols:
    info_wage_median(df_emp, col)

```

### Распределение зарплат

#### **[task 11, 1.5б.]**

Добавить в переменную `df_emp` новый столбец `wage_bin`, где в соответствии с зарплатой будет указан диапазон зарплаты в тыс. рублей. Возможные значения диапазон начинаются от 0 и идут с шагом 4. Например, для зарплаты 2000 рублей должен быть указан диапазон (0,4], а для 5000 рублей – (4,10], и т.д.

```python
values = []
w = (df_emp['wage'] / 1000).to_list()

ends = np.arange(4, max(w)+ 4, 4)
starts = np.arange(0, max(w)+ 4, 4)

for j in range(len(w)):  
    for i in range(len(ends)):
        if w[j] > starts[i] and w[j] <= ends[i]:
            row = '('+str(int(starts[i])) + ',' + str(int(ends[i]))+']'
            values.append(row)
            
df_emp['wage_bin'] = values
df_emp
```

#### **[task 12, 1б.]**
Посчитайте долю каждого диапазона зарплат в выборке, запишите результат в виде Pandas.Series в переменную g.

```python
tmp = []
df1 = df_emp.groupby(by=['wage_bin']).count()
df1 = df1.reset_index()


wage_bin = df1['wage_bin'].tolist()

for w in wage_bin:
    tmp.append(int(w[w.rfind(',') + 1 : -1]))
    
df1['tmp'] = tmp
df1 = df1.sort_values( by = ['tmp'])
df1 = df1.drop(columns = ['tmp'])
df1 = df1.set_index('wage_bin')

g = df1['wage'] / df_emp.shape[0]
```

```python
g.plot(kind='bar')
for index, value in enumerate(g):
    plt.text(index, value, str(round(value,2)))
plt.ylabel('frequency')
plt.show()
```

### Gender wage gap

#### **[task 14, 0.5б.]**
Создайте две подвыборки: одну с мужчинами, чья зарплата меньше или равна 100 тыс. руб, и одну с женщинами, чья зарплата также ниже 100 тыс. руб. 

```python
m_under100k = df.loc[(df['sex'] == 'М') & (df['wage'] <= 100000)]
f_under100k = df.loc[(df['sex'] == 'Ж') & (df['wage'] <= 100000)]
```

#### **[task 15, 0.5б]**

Рассмотрите график. Можно ли говорить о значительной разнице в зарплатах между женщинами и мужчинами? 

[Подробнее о феномене](https://en.wikipedia.org/wiki/Gender_pay_gap).

```python
bins=25

f_under100k['wage'].plot(kind='hist', bins=bins, alpha=0.5, label='Ж')
m_under100k['wage'].plot(kind='hist', bins=bins, alpha=0.5, label='М')

maxx=100000
plt.xlim(0,maxx)

plt.legend()
plt.show()
```

Разница есть, видно, что средняя зарплата у жещин около 20000р, а у мужчин: 30000р.
